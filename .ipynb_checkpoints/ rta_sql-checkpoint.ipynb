# Analyse SQL - Accidents de la Route

```python
import pandas as pd
import sqlite3

print("outils importes")
```

```python
df = pd.read_csv('rta_debutant.csv')
conn = sqlite3.connect('rta_database.db')
df.to_sql('accidents', conn, if_exists='replace', index=False)
print("base de donnees creee")
```

```python
# fonction pour ecrire du sql
def sql(requete):
    return pd.read_sql_query(requete, conn)
```

```python
# voir les 5 premieres lignes
sql("SELECT * FROM accidents LIMIT 5")
```

```python
# table 1 : infos sur l'accident
accidents = df[['Time', 'Day_of_week', 'Cause_of_accident', 'Accident_severity', 'Number_of_casualties', 'Number_of_vehicles_involved']]
accidents.index.name = 'accident_id'
accidents.to_sql('accidents', conn, if_exists='replace', index=True)

# table 2 : infos sur la route
routes = df[['Road_surface_type', 'Road_surface_conditions', 'Road_allignment', 'Weather_conditions', 'Light_conditions']]
routes.index.name = 'accident_id'
routes.to_sql('routes', conn, if_exists='replace', index=True)

print("tables creees")
```

```python
# jointure entre accidents et routes
sql("""
SELECT a.Day_of_week, a.Cause_of_accident, a.Accident_severity,
       r.Road_surface_type, r.Weather_conditions
FROM accidents a
JOIN routes r ON a.accident_id = r.accident_id
LIMIT 10
""")
```

```python
# gravite des accidents selon le type de route
sql("""
SELECT r.Road_surface_type,
       COUNT(*) AS total,
       SUM(CASE WHEN a.Accident_severity = 'Fatal injury' THEN 1 ELSE 0 END) AS mortels
FROM accidents a
JOIN routes r ON a.accident_id = r.accident_id
GROUP BY r.Road_surface_type
ORDER BY mortels DESC
""")
```

```python
# accidents mortels par conditions meteo
sql("""
SELECT r.Weather_conditions,
       COUNT(*) AS total,
       SUM(CASE WHEN a.Accident_severity = 'Fatal injury' THEN 1 ELSE 0 END) AS mortels
FROM accidents a
JOIN routes r ON a.accident_id = r.accident_id
GROUP BY r.Weather_conditions
ORDER BY mortels DESC
""")
```

```python
# jour le plus dangereux avec conditions de route
sql("""
SELECT a.Day_of_week,
       r.Road_surface_conditions,
       COUNT(*) AS total_accidents
FROM accidents a
JOIN routes r ON a.accident_id = r.accident_id
GROUP BY a.Day_of_week, r.Road_surface_conditions
ORDER BY total_accidents DESC
LIMIT 10
""")
```

```python
# fermer la connexion
conn.close()
print("connexion fermee")
```
